```python
import scipy.misc as misc
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.cm as cm
%matplotlib inline
```

```python
# gt = misc.imread('logs/images/gt_1.png')
# inp = misc.imread('logs/images/inp_1.png')
# pred = misc.imread('logs/images/pred_1.png')
num=6
for i in range(9):
    gt = misc.imread('../gt_'+str(i)+'.png')
    inp = misc.imread('../inp_'+str(i)+'.png')
    pred = misc.imread('../pred_'+str(i)+'.png')
    plt.figure(figsize=(50,50))
    i=i+1
    plt.subplot(i*100+31)
    plt.imshow(inp)
    plt.subplot(i*100+32)
    plt.imshow(gt, cmap=cm.Paired ,vmin=0)
    plt.subplot(i*100+33)
    plt.imshow(pred,cmap=cm.Paired ,vmin=0)
```
